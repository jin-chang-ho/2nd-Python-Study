```python
def check(number):
    if number == 2:
        return True
    for i in range(2, number):
        if number % i == 0:
            return False
    return True

lis = []

for month in range(1, 13):
    if month <= 7:
        if month % 2 == 1:
            for i in range (1, 32):
                want = int(str(month) + str(i))
                if check(want):
                    lis = lis + [want]
        elif month == 2:
            for i in range(1, 29):
                want = int(str(month) + str(i))
                if check(want):
                    lis = lis + [want]
        else:
            for i in range(1, 31):
                want = int(str(month) + str(i))
                if check(want):
                    lis = lis + [want]
    else:
        if month % 2 == 0:
            for i in range(1, 32):
                want = int(str(month) + str(i))
                if check(want):
                    lis = lis + [want]
        else:
            for i in range(1,31):
                want = int(str(month) + str(i))
                if check(want):
                    lis = lis + [want]
i = 0
while True:
    if i >= len(lis):
        break
    j = i + 1
    while True:
        if j >= len(lis):
            break
        if lis[i] == lis[j]:
            lis.pop(j)
        j = j + 1
    i = i + 1

for i in lis:
    print(i)
```
